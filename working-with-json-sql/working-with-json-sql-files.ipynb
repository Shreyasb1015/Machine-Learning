```python
import mysql.connector
```

##### To connect our python code to my database we use mysql.connector.connect(). It takes 4 arguments , first is the host where we need to specify the ip address of machine where our database is stored.In our case ,it will be set to localhost ,coz the database is hosted by us locally.Second arg is user where we need to specify our username ,which is by default set to root in the xammp software.Third arg is password ...which we havent set yet to our database ,so we can pass a empty string.Last arg is the database where we need to specify the name of database.

```python
conn=mysql.connector.connect(host='localhost',user='root',password='',database='world')
```

##### To create a dataframe from the sql data, we use pd.read_sql_query().It takes two args ,one is the sql query where we need to specify which columns and which data should be selected and the second arg is connection object ,which in our case is conn.

```python
pd.read_sql_query("SELECT * FROM city",conn)
```

##### We can filter the data as we like using appropriate sql query.

```python
pd.read_sql_query("SELECT * FROM city WHERE CountryCode Like 'IND' ",conn)
```

```python
pd.read_sql_query("SELECT * FROM country",conn)
```
