#### To open a csv file using url

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

pd.read_csv(data)
```

#### To open/load tsv file, we use read_csv().But it is necessary to specify one argument along with it i.e sep which is set to '\t' coz tsv means tab seperated values.The default behaviour of read_csv() is designed to open csv files, and hence sep has default value =',' .But in the case of tsv file ,we should change it to \t.If we dont pass sep argument, then dataset that we load will be in  highly inappropriate manner.

```python
tsv_file=pd.read_csv('movie_titles_metadata.tsv',sep='\t')
```

```python
tsv_file
```

#### But as we can see,the column names are highly incorrect and the first row i.e 0th row is shifted above which represents the column names.Such cases are normal,coz in out tsv file.we werent having names of columns ,thats why this issue is generated.To handle this issue ,we can pass a argument i.e name which will contains the list of names of columns.

```python
tsv_file=pd.read_csv('movie_titles_metadata.tsv',sep='\t',names=['sno','name','release_year','rating','votes','genres'])
```

```python
tsv_file
```

#### As we can observe above that , the dataset which is a csv file is not correctly ordered.The columns names row is shifted one step down.To handle such issue,we can use a argument called header which is set to 1 ,to make 1st row the col names.

```python
pd.read_csv('test.csv',header=1)
```

#### Sometimes .we dont need entire dataset for machine learning purpose ,we just need some of the columns from entire table.To select specified cols while loading the data,we use use_cols argument which takes a list of names of cols that we want.

```python
aug_data=pd.read_csv('aug_train.csv')
```

```python
aug_data.head()
```

```python
pd.read_csv('aug_train.csv',usecols=['enrollee_id','gender','education_level','experience'])
```

#### Skiprows Parameter

#### We can use skiprows parameter to skip specific rows ,which we need to specify using a list.

```python
pd.read_csv('aug_train.csv').head()
```

```python
pd.read_csv('aug_train.csv',skiprows=[0,1]).head()
```

#### The above parameter can be also used in different way ,i.e by calling a func which specifies a list of rows that we must keep.

#### nrows Parameter

#### To get only specified number of rows ,we can use nrows argument and specify how much rows we want in our dataframe.

```python
pd.read_csv('aug_train.csv')
```

```python
pd.read_csv('aug_train.csv',nrows=100)
```

#### Encoding Parameter

##### Usually all the datasets are enoded in utf-8 format ,but some datasets which contains emojis,different language data are not encoded in utf-8 format.To deal with such issues,we can use encoding parameter and then specify our encoding format.

```python
pd.read_csv('zomato.csv',encoding='latin-1')
```
