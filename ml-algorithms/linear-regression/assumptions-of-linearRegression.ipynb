##### Mostly, there are 5 assumptions of linear regression:-
##### 1) Linear Relationship between input and output
##### 2)Multicollinearity
##### 3)Normality of Residual
##### 4)Homoscedasticity
##### 5)Autocorrelation of Residuals


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
X = df.iloc[:,0:3].values
y = df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()

model.fit(X_train,y_train)
```

```python
# Residual
y_pred = model.predict(X_test)
residual = y_test - y_pred
```

### 1. Linear Relationship

```python
fig, (ax1, ax2, ax3) = plt.subplots(ncols=3, figsize=(12, 2.5))

ax1.scatter(df['feature1'], df['target'])
ax1.set_title("Feature1")
ax2.scatter(df['feature2'], df['target'])
ax2.set_title("Feature2")
ax3.scatter(df['feature3'], df['target'])
ax3.set_title("Feature3")

plt.show()
```

### 2. Multicollinearity

##### Multicollinearity occurs when increasing or decreasing values of one column affects other input cols.,ie input cols are related to each other.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif = []

for i in range(X_train.shape[1]):
    vif.append(variance_inflation_factor(X_train, i))
```

```python
pd.DataFrame({'vif': vif}, index=df.columns[0:3]).T
```

##### If the values obtained are 1 or in range of it , then our data doesnt have prblem of multicolinearity.If these values are 5 or more than 5,then our dataset have the prblem of multicolinearity.

```python
# Another Technique
sns.heatmap(df.iloc[:,0:3].corr(),annot=True)
```

### 3. Normality of Residual

```python
sns.displot(residual,kind='kde')
```

##### Residual means the error. It can be calculated by subtracting the original values of y from predicted values.Linear regression says that whenever we plot this residual ,it should be normal.

```python
# QQ Plot

import scipy as sp

fig, ax = plt.subplots(figsize=(6,4))
sp.stats.probplot(residual, plot=ax, fit=True)

plt.show()
```

### 4. Homoscedasticity

##### Homoscedasticity is the pattern obtained when we plot the residual.In homoscedasticity,when we plot the residual, the plot obtained by us should have spread equal.

```python
plt.scatter(y_pred,residual)
```

### 5. Autocorrelation of Residuals

##### It says when we plot the errors/residuals , we should not get any pattern .

```python
plt.plot(residual)
```
