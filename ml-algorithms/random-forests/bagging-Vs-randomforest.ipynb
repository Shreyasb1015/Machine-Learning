```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.ensemble import BaggingClassifier,RandomForestClassifier
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
X,y = make_classification(n_features=5, n_redundant=0, n_informative=5,n_clusters_per_class=1)
```

```python
df = pd.DataFrame(X,columns=['col1','col2','col3','col4','col5'])
df['target'] = y
print(df.shape)
df.head()
```

##### The basic differnce between randomforest and bagging is that in random forest ,we are limited to decision tree models whereas in bagging we can use different models.

##### Another main difference between randomforest and bagging is that in random forest  column sampling occurs at node level i.e when every split occurs , new cols are fetched , whereas in bagging this wont happen and in bagging sampling occurs at tree level.

```python
bag = BaggingClassifier(max_features=2)
```

```python
bag.fit(df.iloc[:,:5],df.iloc[:,-1])
```

```python
plt.figure(figsize=(12,12))
plot_tree(bag.estimators_[0])
plt.show()
```

```python
rf = RandomForestClassifier(max_features=2)
```

```python
rf.fit(df.iloc[:,:5],df.iloc[:,-1])
```

```python

plt.figure(figsize=(12,12))
plot_tree(rf.estimators_[4])
plt.show()
```
