##### Polynomial Logistic regression is used when the data is non -linear

##### For polynomial regression,it is mandatory to select the optimal values of order ,as if we selected higher no of order , the model can overfit and if smaller , the model can underfit.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('ushape.csv')
```

```python
df.head()
```

##### we can rename the columns name using rename func of pandas , for that we must create a dictionary which has keys referring to old column name and values referring to new column name.

```python
new_columns = {'3.159499363321345566e-02': 'X', '9.869877579082642072e-01': 'Y', '0.000000000000000000e+00': 'Class'}
```

```python
df = df.rename(columns=new_columns)
```

```python
df.head()
```

```python
X = df.iloc[:,0:2].values
y = df.iloc[:,-1].values
```

```python
plt.scatter(X[:,0],X[:,1],c=y)
```

##### First , we will see the results obtained by applying logistic regression directly on non-linear dataset.

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression()
```

```python
clf.fit(X,y)
```

```python
from mlxtend.plotting import plot_decision_regions

plot_decision_regions(X, y.astype('int'), clf, legend=2)
```

```python
from sklearn.model_selection import cross_val_score
np.mean(cross_val_score(clf,X,y,scoring='accuracy',cv=10))
```

##### Now , we will perform polynomial regression on input columns.

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=3,include_bias=False)
X_trf = poly.fit_transform(X)
```

```python
clf1 = LogisticRegression()
np.mean(cross_val_score(clf1,X_trf,y,scoring='accuracy',cv=10))
```

```python
def plot_decision_boundary(X,y,degree=1):
    
    poly = PolynomialFeatures(degree=degree)
    X_trf = poly.fit_transform(X)
    
    clf = LogisticRegression()
    clf.fit(X_trf,y)
    
    accuracy = np.mean(cross_val_score(clf,X_trf,y,scoring='accuracy',cv=10))
    
    a=np.arange(start=X[:,0].min()-1, stop=X[:,0].max()+1, step=0.01)
    b=np.arange(start=X[:,1].min()-1, stop=X[:,1].max()+1, step=0.01)


    XX,YY=np.meshgrid(a,b)
    
    input_array=np.array([XX.ravel(),YY.ravel()]).T

    labels=clf.predict(poly.transform(input_array))
    
    plt.contourf(XX,YY,labels.reshape(XX.shape),alpha=0.5)
    plt.scatter(X[:,0],X[:,1], c=y)
    plt.title('Degree = {}, accuracy is {}'.format(degree,np.round(accuracy,4)))
```

```python
plot_decision_boundary(X,y)
```

```python
plot_decision_boundary(X,y,degree=2)
```

```python
plot_decision_boundary(X,y,degree=3)
```

```python
plot_decision_boundary(X,y,degree=4)
```

```python
plot_decision_boundary(X,y,degree=5)
```

```python
plot_decision_boundary(X,y,degree=7)
```
