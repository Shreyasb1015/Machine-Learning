##### In logistic regression , we use general eqn of line i.e Ax+By+C=0 i.e W0X0+W1X1+W3X3....=0

```python
from sklearn.datasets import make_classification
import numpy as np
X, y = make_classification(n_samples=100, n_features=2, n_informative=1,n_redundant=0,
                           n_classes=2, n_clusters_per_class=1, random_state=41,hypercube=False,class_sep=10)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
```

```python
def perceptron(X,y):
    
    #inserting new column in X asnd filling them with 1
    X = np.insert(X,0,1,axis=1)
    ##Creating a array of cols with all initializing to 1
    weights = np.ones(X.shape[1])
    #Specifying learning rate
    lr = 0.1
    
    for i in range(1000):
        j = np.random.randint(0,100)
        #Finding our the model prediction
        y_hat = step(np.dot(X[j],weights))
        #Applying main formula
        weights = weights + lr*(y[j]-y_hat)*X[j]
        
    return weights[0],weights[1:]
```

```python
def step(z):
    return 1 if z>0 else 0
```

```python
intercept_,coef_ = perceptron(X,y)
```

```python
print(coef_)
print(intercept_)
```

##### We can calculate the value of m and b for eqn of line y=mx+c with the help of value of A,B,C by formula m=-A/B  and b=-C/D

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```

##### Creating animation of logistic regression

```python
def perceptron(X,y):
    
    m = []
    b = []
    
    X = np.insert(X,0,1,axis=1)
    weights = np.ones(X.shape[1])
    lr = 0.1
    
    for i in range(200):
        j = np.random.randint(0,100)
        y_hat = step(np.dot(X[j],weights))
        weights = weights + lr*(y[j]-y_hat)*X[j]
        
        m.append(-(weights[1]/weights[2]))
        b.append(-(weights[0]/weights[2]))
        
    return m,b
```

```python
m,b = perceptron(X,y)
```

```python
%matplotlib inline
from matplotlib.animation import FuncAnimation
import matplotlib.animation as animation
```

```python
fig, ax = plt.subplots(figsize=(9,5))

x_i = np.arange(-3, 3, 0.1)
y_i = x_i*m[0] +b[0]
ax.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
line, = ax.plot(x_i, x_i*m[0] +b[0] , 'r-', linewidth=2)
plt.ylim(-3,3)
def update(i):
    label = 'epoch {0}'.format(i + 1)
    line.set_ydata(x_i*m[i] + b[i])
    ax.set_xlabel(label)
    # return line, ax

anim = FuncAnimation(fig, update, repeat=True, frames=200, interval=100)
```

##### Now ,we will use logisticRegression class of sklearn and try to find the difference between our line drawn by perceptron method and the sklearn line drawn by other method.

```python
from sklearn.linear_model import LogisticRegression
lor = LogisticRegression()
lor.fit(X,y)
```

```python
m = -(lor.coef_[0][0]/lor.coef_[0][1])
b = -(lor.intercept_/lor.coef_[0][1])
```

```python
x_input1 = np.linspace(-3,3,100)
y_input1 = m*x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.plot(x_input1,y_input1,color='black',linewidth=3)
plt.scatter(X[:,0],X[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
