##### Shell weight and shucked weight has most no of outliers

##### Both cols are skewed and not normal ,hence we need ot use IQR method.

##### First we will cap the outliers in shell wieght column.

```python
percentile25 = df['Shell weight'].quantile(0.25)
percentile75 = df['Shell weight'].quantile(0.75)
```

```python
percentile25,percentile75
```

```python
iqr=percentile75-percentile25
iqr
```

```python
upper_limit = percentile75 + 1.5 * iqr
lower_limit = percentile25 - 1.5 * iqr
```

```python
print("Upper limit",upper_limit)
print("Lower limit",lower_limit)
```

```python
df[df['Shell weight'] < lower_limit ]
```

```python
df[df['Shell weight'] > upper_limit ]
```

```python
df['Shell weight'] = np.where(
     df['Shell weight'] > upper_limit,
    upper_limit,
    np.where(
          df['Shell weight'] < lower_limit,
           lower_limit,
          df['Shell weight']
    )
)
```

```python
sns.boxplot(df['Shell weight'])
```

##### Now ,we will perform capping on Shucked Weight.

```python
percentile25 = df['Shucked weight'].quantile(0.25)
percentile75 = df['Shucked weight'].quantile(0.75)
```

```python
percentile25,percentile75
```

```python
## Finding iqr
iqr=percentile75-percentile25
iqr
```

```python
upper_limit = percentile75 + 1.5 * iqr
lower_limit = percentile25 - 1.5 * iqr
```

```python
print("Upper limit",upper_limit)
print("Lower limit",lower_limit)
```

```python
df[df['Shucked weight'] > upper_limit]
```

```python
df[df['Shucked weight'] < lower_limit]
```

```python
df['Shucked weight'] = np.where(
     df['Shucked weight'] > upper_limit,
    upper_limit,
    np.where(
          df['Shucked weight'] < lower_limit,
           lower_limit,
          df['Shucked weight']
    )
)
```

```python
sns.boxplot(df['Shucked weight'])
```

```python
X=df.drop(columns='Rings')
y = df['Rings']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

##### Handling categorical column and also scaling.

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import MinMaxScaler

clf=ColumnTransformer(transformers=[
      ('one hot encoding',OneHotEncoder(sparse=False,drop='first'),[0]),
      ('scaling',MinMaxScaler(),slice(1,8)),
])
```

```python
X_train=clf.fit_transform(X_train)
X_test=clf.transform(X_test)
```

#### Training model.

```python
from sklearn.linear_model import LinearRegression

lr=LinearRegression()
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
print("MAE",mean_absolute_error(y_test,y_pred))
print("MSE",mean_squared_error(y_test,y_pred))
print("RMSE",np.sqrt(mean_squared_error(y_test,y_pred)))
print("R2 Score",r2_score(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(max_depth=10,min_samples_leaf=8)
rf.fit(X_train, y_train)
```

```python
y_pred = rf.predict(X_test)
```

```python
print("MAE",mean_absolute_error(y_test,y_pred))
print("MSE",mean_squared_error(y_test,y_pred))
print("RMSE",np.sqrt(mean_squared_error(y_test,y_pred)))
print("R2 Score",r2_score(y_test,y_pred))
```
