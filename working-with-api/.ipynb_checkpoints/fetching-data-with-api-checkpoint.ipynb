##### To work with apis we will need two libraries i.e pandas to convert the data into dataframe and second is request.

```python
import pandas as pd
import requests
```

##### To send a request on the url ,we shoudl use get function of requests library.

```python
response=requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=521a7c68e868e5a699356dbaa962a508&language=en-US&page=1')
```

##### To convert the data i.e stored in response variable into json ,we use json().

```python
response.json()
```

##### We only need the result list from the above json response ,so to extract result list from the above json format which is similar to dictionary ,we should do as follows -->

```python
response.json()['results']
```

##### Now ,we can convert this list into dataframe of pandas.To do so we use DataFrame() of pandas and passes a list as argument.This will convert that list into dataframe.

```python
pd.DataFrame(response.json()['results'])
```

##### Now,it is easy to fetch the columns that we need for work .

```python
df=pd.DataFrame(response.json()['results'])[['id','title','release_date','overview','popularity','vote_average','vote_count']]
```

```python
df
```

##### Now ,we are doing this work for only page no 1,there are more pages in the data fetched by the api.Now to deal with this situtaion ,we will use a loop ,that will run same no of times as the no of pages.For every page,we will create a dataframe and then append that dataframe to the previous dataframe untill loop terminates.While passing a range ,+1 should be added coz loop is exclusive for last index of range.For e.g if we specify range 1 to 429 ,1to 428 pages will be considered.

##### To create a empty dataframe ,  we can use DataFrame() func and pass nothing.

```python
#### pd.DataFrame()
```

##### We will create a empty dataframe ,so that we can append the new dataframes in it.

```python
df=pd.DataFrame()
```

```python
df
```

##### For each page ,will fetch the info and then fetch only that columns which we require and then we will append that dataframe to the empty dataframe created above using concat().It takes 2 arguments, lis of dataframes  and second ignore_index which we will set to True.If we set ignore index to False ,it will first add the data from first iteration from id 0 to let it be 100.Then for next iteration it will again start indexing from 0,which will look bad in the main dataframe.So we should set it to true.

```python
for i in range(1,429):
    response=requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=521a7c68e868e5a699356dbaa962a508&language=en-US&page={}'.format(i))
    temp_df=pd.DataFrame(response.json()['results'])[['id','title','release_date','overview','popularity','vote_average','vote_count']]
    df=pd.concat([df, temp_df ], ignore_index=True) 
```

```python
df
```

##### We can also convert this dataframe into csv file using to_csv() and passing the name of file as argument.

```python
df.to_csv('movies.csv')
```
