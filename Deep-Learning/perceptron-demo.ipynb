```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('placement.csv')
```

```python
print(df.shape)
df.head()
```

```python
sns.scatterplot(df,x='cgpa',y='resume_score',hue='placed')
```

```python
X = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

##### Perceptron is a linear algorithm whose design is such that it went on to become building block of neural network.It is binary classifier,but it can't work on non-linear data.It can take many inputs.Each input as their weights.It also has its bias.Summation function is used to calculate net input function i.e X1w1+X2w2+X3w3+....+b.Here b is bias.After calculating this value ,we passed it to activation function.Activation function is used to bring the input in specific range.Famous activation functions are relu,step function,etc.The eqn of perceptron is just a straight line.

```python
img=plt.imread('perceptron.png')
plt.imshow(img)
```

##### We have inbuilt class of Perceptron in sklearn module.

```python
from sklearn.linear_model import Perceptron
p = Perceptron()
```

```python
p.fit(X,y)
```

```python
p.coef_
```

##### This above coeff are the values of weights.

```python
p.intercept_
```

##### This above intercept is the value of bias.



```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X.values, y.values,clf=p, legend=2)
```
