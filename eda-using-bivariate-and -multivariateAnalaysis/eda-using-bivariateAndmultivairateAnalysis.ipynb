## 1. Scatterplot (Numerical - Numerical)

##### we have scatterplot() of seaborn library to study the relation between two numerical columns.It takes 3 argument,i.e dataframe and the col which should be shown on X- axis and third is the col which should be shown on y axis.

```python
sns.scatterplot(tips,x='total_bill',y='tip')
```

##### For multivariate analysis, which can also differentiate the above plot using the basis of gender using hue parameter

```python
sns.scatterplot(tips,x='total_bill',y='tip',hue=tips['sex'])
```

##### We can also introduce 4th col to differentiate which customers are smoker or not ,using style argument.

```python
sns.scatterplot(tips,x='total_bill',y='tip',hue=tips['sex'],style=tips['smoker'])
```

##### We can also provide one more argumnet i.e size assigned to other column.

```python
sns.scatterplot(tips,x='total_bill',y='tip',hue=tips['sex'],style=tips['smoker'],size=tips['size'])
```

## 2. Bar Plot (Numerical - Categorical)

##### For barplot, we usually placed categorical data on X-axis and numerical data on Y axis. Lets say we want to calculate the average number of people in each Pclass in titanic dataset.

```python
sns.barplot(titanic,x='Pclass',y='Age')
```

```python
sns.barplot(titanic,x='Pclass',y='Fare')
```

##### So ,the above plot shows that class 1 people has paid more fair then class 2 and class 3.

##### For multivariate analysis ,to check how was the fare for each class on the basis of gender ,we can use hue parameter.

```python
sns.barplot(titanic,x='Pclass',y='Fare',hue=titanic['Sex'])
```

```python
sns.barplot(titanic,x='Pclass',y='Age',hue=titanic['Sex'])
```
