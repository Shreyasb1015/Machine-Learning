```python
tips=sns.load_dataset('tips')
```

## 1. Scatterplot (Numerical - Numerical)

##### we have scatterplot() of seaborn library to study the relation between two numerical columns.It takes 3 argument,i.e dataframe and the col which should be shown on X- axis and third is the col which should be shown on y axis.

```python
sns.scatterplot(tips,x='total_bill',y='tip')
```

##### For multivariate analysis, which can also differentiate the above plot using the basis of gender using hue parameter

```python
sns.scatterplot(tips,x='total_bill',y='tip',hue=tips['sex'])
```

##### We can also introduce 4th col to differentiate which customers are smoker or not ,using style argument.

```python
sns.scatterplot(tips,x='total_bill',y='tip',hue=tips['sex'],style=tips['smoker'])
```

##### We can also provide one more argumnet i.e size assigned to other column.

```python
sns.scatterplot(tips,x='total_bill',y='tip',hue=tips['sex'],style=tips['smoker'],size=tips['size'])
```

## 2. Bar Plot (Numerical - Categorical)

##### For barplot, we usually placed categorical data on X-axis and numerical data on Y axis. Lets say we want to calculate the average number of people in each Pclass in titanic dataset.

```python
sns.barplot(titanic,x='Pclass',y='Age')
```

```python
sns.barplot(titanic,x='Pclass',y='Fare')
```

##### So ,the above plot shows that class 1 people has paid more fair then class 2 and class 3.

##### For multivariate analysis ,to check how was the fare for each class on the basis of gender ,we can use hue parameter.

```python
sns.barplot(titanic,x='Pclass',y='Fare',hue=titanic['Sex'])
```

```python
sns.barplot(titanic,x='Pclass',y='Age',hue=titanic['Sex'])
```

## 3. Box Plot (Numerical - Categorical)

##### We can  also do  bivariate analysis using boxplot() to check the outliers of female and male.

```python
sns.boxplot(titanic,x='Sex',y='Age')
```

##### For multivariate analysis ,we can use hue parameter .Lets find the no of people who survived on the basis of age and gender.

```python
sns.boxplot(titanic,x='Sex',y='Age',hue=titanic['Survived'])
```

##  Histplot (Numerical - Categorical)

##### Lets plot histplot for those people who cannot survived on the basis of age.

```python
sns.histplot(titanic[titanic['Survived']==0]['Age'])
sns.histplot(titanic[titanic['Survived']==1]['Age'])
```

## 5. HeatMap (Categorical - Categorical)

##### To get a basic small grid between two columns, we can us pd.crosstab()  and pass two cols.

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
```

##### To plot this grid ,we can use heatmap().

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

##### To plot the number of people survived in titanic using percenatge ,we can follow the below steps:

```python
titanic.groupby('Pclass').mean('Survived')['Survived']*100
```

```python
(titanic.groupby('Pclass').mean('Survived')['Survived']*100).plot(kind='bar')
```

```python
(titanic.groupby('Sex').mean('Survived')['Survived']*100).plot(kind='bar')
```

## 6. ClusterMap (Categorical - Categorical)

```python
## It is similar to heatmap ,but it shows the closeness between values.
```

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
```

## 7. Pairplot

##### Pairplot is similar to scatterplot func but even more useful.It automatically identifies numerical columns in our dataframe and plot scatterplot for each col vs each col by itself.

```python
sns.pairplot(iris)
```

##### For same col vs same col plot,it has plot histogram instead of scatterplot.

##### For multivariate analysis ,we have hue paramter and can pass another paraemeter which is categorical

```python
sns.pairplot(iris,hue='species')
```

## 8. Lineplot (Numerical - Numerical)

##### LinePlot is mainly used when on our X-axis ,we have time based func like time ,year,month,day,week,etc

##### If we want to plot the lineplot for no of people travelled every year, we can do it as following -->

```python
flights.groupby('year').sum('passengers').reset_index()
```

```python
new=flights.groupby('year').sum('passengers').reset_index()
```

```python
sns.lineplot(x=new['year'],y=new['passengers'])
```

##### Trying heatmap on flights dataframe.

```python
sns.heatmap(flights.pivot_table(values='passengers',index='month',columns='year'))
```
