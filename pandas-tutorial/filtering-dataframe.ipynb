```python
import numpy as np
import pandas as pd
```

```python
data=pd.read_csv('matches - matches.csv')
```

```python
data.head()
```

##### Now we will look to filter the data according to the condition.This process is called masking.Here we create a variable called mask.We then fetch any required column using condition . It returns a series of true and false i.e boolean series. For e.g if I want to fetch only that rows where column 'city' has value equal to hyderabad,we can do it as follows:

```python
mask=data['city'] =='Hyderabad'
data[mask]

# The above command means that it will print only those rows whose mask value is true,not to forget mask is the boolean series which contains either
#true or false value
```

##### To get the total no of cols and rows from the above new obtained dataframe ,we can use shape attribute.|

```python
data[mask].shape
```

##### To get only the no of columns in the dataframe ,we can use indexing that is 0 index for number of cols.

```python
data[mask].shape[0]
```

##### We can also define the func to find the total no of matches in the specific city where user gives us cityname

```python
def get_city_match_count(city):
    mask=data['city']==city
    return data[mask].shape[0]
```

```python
get_city_match_count('Hyderabad')
```

```python
get_city_match_count('Pune')
```

##### We can also some complex conditions while learning/training which includes 2 or more conditions.In this case we create more number of masks.For each condition,we create one mask.For e.g we I want the count of matches that happened in hyderabad after 2010 ,to fetch such rows ,we need to use two masks

```python
mask1=data['city']=='Hyderabad'
mask2=data['date']>'2010-01-01'
```

```python
data[mask1 & mask2]

# & is used instead of 'and' coz , we are performing bitwise operation on boolean values ,hence & operator is used.
```

##### To get the count of above macthes,we can use shape attribute and  extract the no of rows only in it.

```python
data[mask1 & mask2].shape[0]
```

##### Now we are going to use one useful fucntion i.e value_counts(). This function is applied to only categorical data or series.First we need to fetch the series /column where we want to apply this function and then we will apply this function to that series.For e.g if we want to find out how many number of matches does the each team in tournament won ,then we can do the following :-

```python
data['winner'].value_counts()
```

##### Now we are going to look pandas plot function

##### Pandas plot function can draw different kind of data like line ,bar,area,pie,box,etc.Now we will plot the above info i.e of count of matches wan by each team in the form of bar chart.Bar chart has categorical value on X axis while numerical value on Y axis. For that ,we need to import pyplot from matplotlib so that graph becomes visible on the screen.

```python
import matplotlib.pyplot as  plt
```

```python
data['winner'].value_counts().plot(kind='bar')
```

##### To get only top 5 teams ,we can use head().

```python
data['winner'].value_counts().head().plot(kind='bar')
```
