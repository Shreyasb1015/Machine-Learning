```python
import numpy as np
import pandas as pd
```

```python
company=pd.read_csv('Fortune 500 2017 - Fortune 500.csv')
company.head()
```

##### Now we are going to use one imp func for which is used for analyzing data i.e groupby().This func is mainly used on categorical data.From the above table ,there are two such columns which seems to be logically where we can group our data i.e on the basis of sector and industry.

```python
company.groupby('Sector')
```

##### From the above output ,we can see that output is not exciting as we expected.But we will store this groupby object in variable called sector and will use this variable for many purposes which are extremely useful.

```python
sectors=company.groupby('Sector')
```

```python
sectors
```

##### We can use len() to know how much different sectors do we have in the sectors variable which is obtained by groupby().

```python
len(sectors)
```

##### We can also get the no of companies which are in the various sector by using size().

```python
sectors.size()
```

##### We can also sort the values using sort_values() in descending order to know which sector has most no of companies.

```python
sectors.size().sort_values(ascending=False)
```

##### We can get the info/row of  first company of each sector using first().

```python
sectors.first()
```

##### We can also get the last company for each sector in the dataframe using last()

```python
sectors.last()
```

##### Another imp attribute is groups .This attribute returns a dictionary where key is the unique value in the column where we applied groupby() (in this case key is the sector name) and value is the id of different companies.

```python
sectors.groups
```

```python
### We can cross-check the above output by fetching 23rd row and all its columns.We must find that it comes under Aerospace & Defense sector.
```

```python
company.iloc[23, :]
```

##### Another useful function is get_group().It takes one argument i.e name of group on the basis we want to classify our groupby object.It returns us a dataframe.For e.g if we want to all the companies that comes under Energy group ,we can implement the following:

```python
sectors.get_group('Energy')
```

##### We can use sum( ) to get the sum of all values of each numerical columns that are in the each sector in sectors dataframe

```python
sectors.sum()
```
